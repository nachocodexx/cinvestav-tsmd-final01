```python
# %matplotlib inline
%matplotlib notebook

def plotClustering(data,**kwargs):
    fig         = plt.figure(figsize=(10,10))
    ax          = fig.add_subplot(111, projection= '3d')
    s           = kwargs.get('size',50 ) 
    alpha       = kwargs.get('alpha',.5)
    colors      = kwargs.get('colors',[
        '#F44336','#673AB7','#607D8B',
#         '#9C27B0',
        'black',
        '#CDDC39','#2196F3',
        '#00BCD4','#009688',
        '#4CAF50','#FFC107','#FF5722','#795548'
    ])
    title       = kwargs.get('title','Principal components')
    axis_labels = kwargs.get("axis_labels",["$\mathbf{P_1}$",'$\mathbf{P_2}$','$\mathbf{P_3}$'])
    for key,value in data.items():
        values = np.array(value)
        ax.scatter(values[:,0],values[:,1],values[:,2],color=colors[int(key)],s=s,alpha=alpha,
                   label="Cluster-{}".format(key))
    ax.set_title(title)
    ax.set_xlabel(axis_labels[0])
    ax.set_ylabel(axis_labels[1])
    ax.set_zlabel(axis_labels[2])
    ax.legend(bbox_to_anchor=(1.1, 1))
    return ax,colors
# data,labels  = applyClustering(p_components,KMeans(n_clusters=4,random_state=0))
```

```python
class Clustering(ABC):
    def __init__(self,**kwargs): 
        self.key              = kwargs.get('key')
        self.fx               = kwargs.get('fx')
        self.ploat_color      = kwargs.get('plot_color')
        self.plot_title       = kwargs.get('plot_title')
        self.data             = kwargs.get('data')
        self.CLUSTERING_NAMES = {'SHS':"Silhouette",'CHS':"Davies Bouldin",'DBS':"Calinski Harabasz"}
    def apply_clustering(self,**kwargs):
        X       = kwargs.get('X',self.data)
        k       = kwargs.get('k',2)
        result  =  self.fx(k).fit(X)
        labels  = result.labels_
        data    = dict()
        for i,xi in enumerate(X):
            cluster = str(labels[i])
            if not(cluster in data):
                data[cluster]=[]
            data[cluster].append(xi)
        scores = {'SHS':SHS(X,labels), 'CHS':CHS(X,labels),'DBS':DBS(X,labels)}
        return (data,labels,scores)
    def apply_clusterings(self,**kwargs):
        ks = kwargs.get("ks",list(range(2,10)))
        scores_keys = ['SHS','CHS','DBS']
        _scores = {'SHS':[],'CHS':[],'DBS':[]}
        for k in ks:
            data,labels,scores = self.apply_clustering(k=k)
            for sk in scores_keys:
                _scores[sk].append(scores[sk])
        return ks,_scores,
    def plot_scores(self,**kwargs):
        ks = kwargs.get('ks',list(range(2,10)))
        _,ax = plt.subplots(3,1)
        ks,scores = self.apply_clusterings(ks=ks)
        for i,(name,score) in enumerate(scores.items()):
            title = self.plot_title+"-"+self.CLUSTERING_NAMES[name]
            ax[i].set_title(title)
            ax[i].plot(ks,score,color=COLORS[i],marker="o")
    def plot(self,**kwargs):
        k           = kwargs.get("k",2 )
        figsize     = kwargs.get("figsize",(10,10))
        _,ax        = plt.subplots(figsize=figsize)
        
        
        
class KMeansClusteringX(Clustering):
    def __init__(self,*args,**kwargs): 
        super().__init__(*args,**kwargs)
class SpectralClusteringX(Clustering):
    def __init__(self,*args,**kwargs): 
        super().__init__(*args,**kwargs)        
class AgglomerativeClusteringX(Clustering):
    def __init__(self,*args,**kwargs): 
        super().__init__(*args,**kwargs)        
        
kmeans = KMeansClusteringX(
                 key        = "kmeans",
                 fx         = lambda k:  KMeans(n_clusters=k,random_state=0),
                 plot_color = "red",
                 plot_title = "K-means", 
                 data       = p_components
)

spectral = SpectralClusteringX(
                 key        = "spectral",
                 fx         = lambda k: SpectralClustering(n_clusters=k,assign_labels='discretize',random_state=0),
                 plot_color = "blue",
                 plot_title = "Spectral Clustering", 
                 data       = p_components
)
agglomerative = AgglomerativeClusteringX(
                 key        = "agglomerative",
                 fx         = lambda k: AgglomerativeClustering(n_clusters=k),
                 plot_color = "green",
                 plot_title = "Agglomerative Clustering", 
                 data       = p_components
)

# spectral.plot()
```

```python
def plot2DClustering(data,**kwargs):
    figsize     = kwargs.get("figsize",(10,10))
    _,ax        = plt.subplots(figsize=figsize)
    s           = kwargs.get('size',20 ) 
    alpha       = kwargs.get('alpha',.9)
    colors      = kwargs.get('colors',[
        '#F44336','#673AB7','#607D8B',
        'black',
        '#CDDC39','#2196F3',
        '#00BCD4','#009688',
        '#4CAF50','#FFC107','#FF5722','#795548'
    ])
    title       = kwargs.get('title','Default Title')
    axis_labels = kwargs.get("axis_labels",["$P_1$",'$P_2$','$P_3$'])
    for key,value in data.items():
        values = np.array(value)
        ax.scatter(values[:,0],values[:,1],color=colors[int(key)],s=s,alpha=alpha,
                   label="Cluster-{}".format(key))
    ax.set_title(title)
    ax.set_xlabel(axis_labels[0])
    ax.set_ylabel(axis_labels[1])
    ax.legend(bbox_to_anchor=(1.3, 1))
    return ax,colors
```

# Kmeans

```python
%matplotlib notebook
def generateCountTable(**kwargs):
    result   = kwargs.get('data',X)
    roles    = kwargs.get('roles',X_roles) 
    labels   = kwargs.get('labels') 
    role_col = kwargs.get('role_col',"USER_ROLE")
    id_col   = kwargs.get('id_col','USER_ID')
    result[role_col] =roles 
    result['CLUSTER']   = labels
    res = result.pivot_table(
        index=["CLUSTER",role_col],
        values=[id_col],
        aggfunc=len
    )
    res.columns=['COUNT']
    return res
def clusteringPCA(**kwargs):
    title        = kwargs.get('plot_title')
    show_out     = kwargs.get('show_out',False)
    output_csv   = kwargs.get('output_csv')
    show_plot    = kwargs.get('show_plot',False)
    k            = kwargs.get('k',2)
    svd_solver   = kwargs.get('svg_solver',0)
    data         = kwargs.get('data',X_scaled)
    clustering   = kwargs.get('clustering')
    pca          = PCA(n_components=3,svd_solver=svd_solvers[svd_solver])
    p_components = pca.fit_transform(data)
#     clustering   = KMeans(n_clusters=k,random_state=0)
    # clustering = SpectralClustering(n_clusters=k,random_state=0,assign_labels='discretize')
    data,labels  = applyClustering(p_components,clustering)
    if(show_plot):
        plotClustering(data,title=title)
    generateCountTable(labels=labels).to_csv("results/{}".format(output_csv))
    if(show_out):
        return data
ks = range(2,10)
for k in ks:
    title      = 'PCA Agglomerative Clustering - k='+str(k)
    output_csv = 'agg_'+str(k)
    clustering = AgglomerativeClustering(n_clusters=k)
    clusteringPCA(data=X_scaled,
                  output_csv=output_csv,
                  plot_title=title,
                  k=k,
                  svd_solver=0,
                  clustering=clustering)
```
